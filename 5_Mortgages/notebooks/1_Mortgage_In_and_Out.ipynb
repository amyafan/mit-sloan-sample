how many people get new mortgages each month? 

```python
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

output_3 = "~//projects//equifaxmacro_proj//EquiFax3//3_Balance_In_And_Out_Construction//output//"
```

```python
cols = ['first_c', 'last_c']

biao_str = output_3 + "1_balance_in_and_out.parquet"

biao = pd.read_parquet(biao_str, columns = cols) 
```

```python
biao = biao.reset_index()
biao = biao[biao.product_category == "FM"]
```

```python
biao.first_c.value_counts()
```

```python
pd.concat([biao.last_c.value_counts(), biao.first_c.value_counts()], axis = 1).sort_index() 
```
