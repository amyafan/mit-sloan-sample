```python
biao_str = output_3 + "1_balance_in_and_out.parquet"

cols = [str(year) + '01' for year in range(2006, 2022)]

iao = pd.read_parquet(biao_str, columns = cols).reset_index()
```

```python
# figuring out the number of people with a mortgage each month 
mort = iao[iao.product_category == "FM"]
```

```python
mort_counts = mort.trade_id.value_counts()
```

```python
# find all the mortgages that show
mort_2 = mort_counts[mort_counts > 1].index
mort_2
```
